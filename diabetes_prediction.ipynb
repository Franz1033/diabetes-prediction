# Diabetes Prediction using Logistic Regression

A machine learning model that predicts if an individual is diabetic or non-diabetic (Binary Classification)

## Introduction

Diabetes is a disease that causes your blood sugar to be too high, this can be categorized in to two types, type 1 diabetes in which the pancreas does not make insulin, because the body's immune system attacks the islet cells in the pancreas that make insulin and type 2 diabetes which the pancreas makes less insulin than used to, and your body becomes resistant to insulin. Alarmingly, it was found that around 10.5% of the adult population in the world has diabetes according to the international Diabetes Federation, with almost half being unaware that they are living with the condition, thats around less than 700 million in terms of the world population. In the Philippines alone, around 4.3 million Filipinos have this disease according to a study in 2021.

This disease is fueled by a myriad of factors such as family history, diet, age, pre-existing medical conditions, and overall lifestyle which makes the exact cause more challenging to determine. Oftentimes, it's a combination of these factors that causes an indivudual to possess diabetes. This made our group to be drawn to explore this domain because not only is this a global cause for concern, but our efforts in this area though little, can potentially create massive benefits to indivduals that may be grappling with this disease. Furthermore, because diabetes is a serious disease that can be fatal if left uncared of, we want to further investigate and deepen our understanding of the fundamental causes of this disease by applying machine learning techniques. Specifically, we'll utilize logistic regression to predict if an individual is diabetic or not based on some given criteria such as age, bmi, smoking habits, glucose levels, and more. Through this analysis, it will enhance early detection and diagnosis of diabetes and will provide significant benefits to the healthcare industry. Which will lead to earlier intervention and prevent further progression of the disease.

## Dataset

The dataset is composed of health and demographic data of 100,000 individuals, intended to facilate diabetes-related research and predictive modeling. It includes information such as gender, age, location, race, hypertension, heart disease, smoking history, BMI, HbA1c level, blood glucose level, and diabetes status.

Kaggle source link: https://tinyurl.com/2yerps5j

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from joblib import dump, load
```

```python
df = pd.read_csv('diabetes_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

## Methodology

Before training the model, we first perform necessary preprocessing steps with our data to ensure correctness and compatibility. Since the machine learning algorithm that'll be used for this analysis doesn't directly handle variables with categorical values especially in its string representation, we apply a common preprocessing technique called one-hot encoding. This technique transforms each unique value of our nominal feature into its own separate column containing a binary representation.

```python
df = pd.get_dummies(df, columns=["gender", "location", "smoking_history"]) # Perform one-hot encoding to given columns
```

```python
df.head() # One-hot encoded DataFrame
```

```python
print(f"Total Variables: {df.columns.size}")
```

We split our dataset into training and test set which will help us in evaluating our model later on.

```python
train_set, test_set = train_test_split(df, test_size=0.2) # Using a 8:2 ratio for train-test split

print(f"Training Set: {len(train_set)} samples (80%)")
print(f"Test Set: {len(test_set)} samples (20%)")
```

We define the input variables and target variable for training and testing

```python
train_X = train_set.copy().drop(columns="diabetes").to_numpy() # Input variables
train_y = train_set.loc[:, ["diabetes"]].to_numpy().ravel() # Target variable

test_X = test_set.copy().drop(columns="diabetes").to_numpy() # Input variables
test_y = test_set.loc[:, ["diabetes"]].to_numpy().ravel() # Target variable
```

We define the logistic regression model, a machine learning algorithm well-suited for binary classification tasks. Which works perfectly in this case as we want to determine if an individual is diabetic or non-diabetic (binary classification). Logistic regression is an appropriate choice for these tasks as it uses a logistic function that outputs values between 0 and 1. The formula is as follows: `f(x) = 1 / (1 + e^-x)` where `x` is our input and `e` is the Euler's number, we also follow the standard threshold of 0.5. This function produces a sigmoid curve, which we use as the basis for our predictions.

```python
model = LogisticRegression(random_state=0, max_iter=2000) # Tuned to ensure convergence when training
```

Finally, we train the model using the corresponding features and target of the training set

```python
model.fit(train_X, train_y) # Fit model to training data
```

This model is evaluated on the next section.

## Results and Analysis

We then evaluate the model using our test set

```python
test_accuracy = model.score(test_X, test_y)

print(f"Test Accuracy: {test_accuracy * 100}%")
```

The test accuracy shows the percentage of predictions that matched the true labels in our test set

To futher evaluate the performance of the algorithm on the dataset, we can use the confusion matrix and the four metrics of Accuracy, Precision, Recall, and the F1-Score.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Predict the labels for the test set
y_pred = model.predict(test_X)
conf_matrix = confusion_matrix(test_y, y_pred)

# Print the confusion matrix
print('Confusion Matrix Results')
print(conf_matrix)

# Confusion Matrix Visualization
print('\t\t\tConfusion Matrix Visualization')
sns.heatmap(conf_matrix, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Assign variables based on the confusion matrix
true_negatives = conf_matrix[0][0]
false_positives = conf_matrix[0][1]
false_negatives = conf_matrix[1][0]
true_positives = conf_matrix[1][1]

# Calculate the metrics
accuracy = (true_positives + true_negatives) / (true_positives + true_negatives + false_positives + false_negatives)
precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
f1_score = (2 * (precision * recall)) / (precision + recall)

# Print the metrics and their interpretations
print(f"True Negatives (TN): {true_negatives} samples were correctly predicted as non-diabetic.")
print(f"False Positives (FP): {false_positives} samples were incorrectly predicted as diabetic (Type I error).")
print(f"False Negatives (FN): {false_negatives} samples were incorrectly predicted as non-diabetic (Type II error).")
print(f"True Positives (TP): {true_positives} samples were correctly predicted as diabetic.")
print()

total_samples = len(test_y)
total_errors = false_positives + false_negatives
correct_predictions = true_negatives + true_positives

print(f"In total, {total_errors} out of {total_samples} samples resulted in an error.")
print(f"Whereas, the remaining {correct_predictions} out of {total_samples} samples had correct predictions.")
print()

# Print the results of the metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1_score:.2f}")

# Interpretations
print("\n1.) Accuracy")
print("Accuracy Score Interpretations:")
print("Greater than 0.9 = Great,")
print("Greater than 0.7 = Decent,")
print("Less than 0.7 = Mediocre")
if accuracy > 0.9:
    print(f"The resulting accuracy score was {accuracy:.2f}. This indicates that the algorithm performed great in correctly classifying the samples for diabetes.")
elif accuracy > 0.7:
    print(f"The resulting accuracy score was {accuracy:.2f}. This indicates that the algorithm performed decently in correctly classifying the samples for diabetes.")
else:
    print(f"The resulting accuracy score was {accuracy:.2f}. This indicates that the algorithm performed mediocrely in correctly classifying the samples for diabetes.")

print("\n2.) Precision")
print(f"The precision score was {precision:.2f}, or {precision*100:.0f}%. This means that out of all the predicted positive samples, {precision*100:.0f}% were actually positive.")

print("\n3.) Recall")
print(f"The recall score was {recall:.2f}, or {recall*100:.0f}%. This indicates that out of all the actual positive samples, {recall*100:.0f}% were correctly predicted.")

print("\n4.) F1-Score")
print("F1-Score Interpretations:")
print("Greater than 0.9 = Very Good,")
print("0.8-0.9 = Good,")
print("0.5-0.8 = OK,")
print("Less than 0.5 = Not Good")
if f1_score > 0.9:
    print(f"The resulting F1-Score was {f1_score:.2f}. This suggests that the algorithm had a very good performance.")
elif f1_score >= 0.8:
    print(f"The resulting F1-Score was {f1_score:.2f}. This suggests that the algorithm had a good performance.")
elif f1_score >= 0.5:
    print(f"The resulting F1-Score was {f1_score:.2f}. This suggests that the algorithm had an OK performance, but there is room for improvement.")
else:
    print(f"The resulting F1-Score was {f1_score:.2f}. This suggests that the algorithm's performance was not good, and significant improvements are needed.")
```
